# Method 1 - using a dictionary

```python
# count the number of words in t8.shakespeare.txt using a dictionary
def count_dictionnary(filename):
    word_count = {}
    with open(filename, 'r') as f:
        for line in f:
            for word in line.split():
                if word in word_count:
                    word_count[word] += 1
                else:
                    word_count[word] = 1
    return word_count

word_count = count_dictionnary('t8.shakespeare.txt')

word_count
```

# Method 2 - using the Counter function

```python
from collections import Counter

def count_counter(filename):
    with open(filename, 'r') as f:
        lines = f.readlines()
    text = ' '.join(lines)
    words = text.split()
    counter = Counter(words)
    return counter

counter = count_counter('t8.shakespeare.txt')
count_counter('t8.shakespeare.txt')
```

# Compare the 2 methods

```python
%timeit count_dictionnary('t8.shakespeare.txt')
%timeit count_counter('t8.shakespeare.txt')
```

```python
# run this experiment 100 times and plot the two distributions of execution times.
import timeit
import numpy as np
import matplotlib.pyplot as plt

def time_dictionnary():
    count_dictionnary('t8.shakespeare.txt')

def time_counter():
    count_counter('t8.shakespeare.txt')

dictionnary_time = np.array(timeit.repeat(time_dictionnary, repeat=100, number=1))
counter_time = np.array(timeit.repeat(time_counter, repeat=100, number=1))

plt.hist(dictionnary_time, bins=20, alpha=0.5, label='Dictionnary')
plt.hist(counter_time, bins=20, alpha=0.5, label='Counter')
plt.legend(loc='upper right')

plt.xlabel('Time')
plt.ylabel('Frequency')

plt.title('Execution times for count_dictionnary and count_counter')
plt.show()
```

```python
# compute the mean and variance of the two distributions
print('dictionnary: mean = {:.3f}, std = {:.3f}'.format(dictionnary_time.mean(), dictionnary_time.std()))
print('counter: mean = {:.3f}, std = {:.3f}'.format(counter_time.mean(), counter_time.std()))
```
